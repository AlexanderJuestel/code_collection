```python
def download_dois_one_issue_mdpi(journal: str,
                                 volume_number: int,
                                 issue_number: int) -> list:
    """Function to download DOI numbers for one MDPI Journal issue
    
    Parameters:
    ___________
    
        journal: str
            Journal identifier
            
        volume_number: int
            Volume number
            
        issue_number: int
            Issue number
            
    """
    
    # Checking that the journal identifier is of type str
    if not isinstance(journal, str):
        raise TypeError('Journal identifier must be provided as string')
    
    # Checking that the volume number is of type int
    if not isinstance(volume_number, (int, np.integer)):
        raise TypeError('Volume Number must be of type int')
        
    # Checking that the issue number is of type int
    if not isinstance(issue_number, (int, np.integer)):
        raise TypeError('Issue Number must be of type int')
        
    journal_dict = {'energies' : '1996-1073'}
        
    # Creating the URL
    url = 'https://www.mdpi.com/%s/%s/%s' % (journal_dict[journal], volume_number, issue_number)
        
    # Getting the website content
    website = doi_downloads.get_website_content(url=url)
    
    # Extracting the DOI Numbers
    dois_mdpi = ['https' + website.split('>https')[i+ 1].split('</a>')[0] for i in range(len(website.split('>https'))-1)][:-1]
    
    return dois_mdpi


def download_dois_multiple_issues_mdpi(journal: str,
                                       volume_number: int,
                                       issue_start: int,
                                       issue_end: int) -> list:
    
    """Function to download DOI numbers for multiple MDPI Journal issues
    
    Parameters:
    ___________
    
        journal: str
            Journal identifier
            
        volume_number: int
            Volume number
            
        issue_start: int
            Issue number start
        
        issue_end: int 
            Issue number end
            
    """
    
    # Checking that the journal identifier is of type str
    if not isinstance(journal, str):
        raise TypeError('Journal identifier must be provided as string')
    
    # Checking that the volume number is of type int
    if not isinstance(volume_number, (int, np.integer)):
        raise TypeError('Volume Number must be of type int')
        
    # Checking that the issue number start is of type int
    if not isinstance(issue_start, int):
        raise TypeError('Issue Number start must be of type int')
        
    # Checking that the issue number end is of type int
    if not isinstance(issue_end, int):
        raise TypeError('Issue Number end must be of type int')
        
    # Creating range of issues
    issues = np.arange(issue_start, issue_end+1, 1)
    
    
    # Creating empty list to store DOI Numbers
    dois_mdpi = []
    
    # Extracting DOI numbers
    for issue in issues:
        dois_single_issue = download_dois_one_issue_mdpi(journal=journal,
                                                         volume_number=volume_number,
                                                         issue_number=issue)
        
        dois_mdpi = dois_mdpi + dois_single_issue
        
    return dois_mdpi

def download_dois_mdpi(journal: str,
                       volume_start: int,
                       volume_end: int,
                       issue_start: int = 1,
                       issue_end: int = 20) -> list:
    
    """Function to download DOI numbers for multiple MDPI Journal volumes
    
    Parameters:
    ___________
    
        journal: str
            Journal identifier
            
        volume_start: int
            Volume number start
            
        volume_end: int
            Volume number end
            
        issue_start: int
            Issue number start
        
        issue_end: int 
            Issue number end
            
    """
    
    # Checking that the journal identifier is of type str
    if not isinstance(journal, str):
        raise TypeError('Journal identifier must be provided as string')
    
    # Checking that the volume number start is of type int
    if not isinstance(volume_start, (int, np.integer)):
        raise TypeError('Volume Number start must be of type int')
        
    # Checking that the volume number end is of type int
    if not isinstance(volume_end, (int, np.integer)):
        raise TypeError('Volume Number end must be of type int')
        
    # Checking that the issue number start is of type int
    if not isinstance(issue_start, int):
        raise TypeError('Issue Number start must be of type int')
        
    # Checking that the issue number end is of type int
    if not isinstance(issue_end, int):
        raise TypeError('Issue Number end must be of type int')
        
    # Creating range of volumes
    volumes = np.arange(volume_start, volume_end+1, 1)
    
    # Creating empty list to store DOI Numbers
    dois_mdpi = []
    
    # Extracting DOI numbers
    for volume in volumes:
        dois_multiple_issues = download_dois_multiple_issues_mdpi(journal=journal,
                                                                  volume_number=volume,
                                                                  issue_start=issue_start,
                                                                  issue_end=issue_end)
        
        dois_mdpi = dois_mdpi + dois_multiple_issues
        
    return dois_mdpi
```
