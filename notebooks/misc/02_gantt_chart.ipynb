# Timeline Promotion

This notebook illustrates how to create a Gantt Chart. 

# Import Libraries

```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

# Load Timeline Data

```python
#Read Data from schedule.csv
df =pd.read_excel('../../data/misc/Timeline_Promotion.xlsx')
df
```

# Converting dates to datetime format

```python
#Convert dates to datetime format
df.start=pd.to_datetime(df.start)
df.end=pd.to_datetime(df.end)
```

# Adding duration

```python
#Add Duration
df['duration']=df.end-df.start
df.duration=df.duration.apply(lambda x: x.days+1)
```

# Sorting in ascending order of start date

```python
#sort in ascending order of start date
df=df.sort_values(by='start', ascending=True)
```

# Projecting Level Variables

```python
#project level variables
p_start=df.start.min()
p_end=df.end.max()
p_duration=(p_end-p_start).days+1
```

# Adding relative date

```python
#Add relative date
df['rel_start']=df.start.apply(lambda x: (x-p_start).days)
```

# Creating custom ticks and labels

```python
#Create custom x-ticks and x-tick labels
x_ticks=[i for i in range(p_duration+1)]
x_labels=[(p_start+dt.timedelta(days=i)).strftime('%d-%b') 
          for i in x_ticks]
```

# Adding Color

```python
c_dict={'Publication':'red', 'Project':'orange', 'Conference':'green'}
```

# Plotting without colors

```python
######  PLOTTING GANTT CHART ######
plt.figure(figsize=(12,8))
plt.title('Gantt Chart Promotion', size=18)
plt.barh(y=df.Task, left=df.rel_start, width=df.duration)
plt.gca().invert_yaxis()
plt.yticks(fontsize=14)
plt.xticks(ticks=xtick, labels=xlabs, rotation = 45)
plt.grid(axis='x')
plt.show()
```

# Plotting with colors

```python
plt.figure(figsize=(12,8))
plt.title('Gantt Chart Promotion', size=18)
for i in range(df.shape[0]):
    color=c_dict[df['Type'][i]]
    plt.barh(y=df.Task[i], left=df.rel_start[i], 
             width=df.duration[i], alpha=0.4, 
             color=color)
    
plt.gca().invert_yaxis()
plt.xticks(ticks=xtick, labels=xlabs, rotation = 45, fontsize=14)
plt.grid(axis='x')
plt.yticks(fontsize=14)
plt.show()
```
